Transformer models are used to predict the probability of a given event occurring in a sequence of events. The probability is calculated by taking the product of the probabilities of each event in the sequence. For example, the event "John is a student" is the first event of an event sequence, and the second event is "George is an employee". If the events are independent, then the model can be used for predicting the next event. However, if the two events have some correlation, such as "john is also a teacher", "george is not a member of staff" or "johns is in charge of finance" and "geroge is responsible for the accounting department" then a more complex model is required. In this case, a Markov chain model (MCM) is used. A MCM is similar to a Bayesian network, but it is more flexible and can handle more events than a BN. MCCs are also used in other fields, for example in financial forecasting. They are often used as a tool to model the correlation between events, which is important for financial prediction. This is because the financial market is highly correlated, so it would be difficult to forecast the future with a single model. Therefore, it makes sense to use a MCCM to improve the accuracy of financial predictions.
The MCCA is based on the MMM, where the transition probabilities are calculated using the conditional probabilities. It is different from the traditional MMC, because it does not use the Markov property. Instead, each transition probability in MCA is determined by the previous transition and current state. Thus, MMCA is able to handle the correlations between the states. As a result, this model has the ability to deal with the complex events in finance. Figure 1 shows the flowchart of MACA. First, we need to define the state space of our model, that is, all possible states of all events that can occur in an MCCC. Then, given the initial state, an algorithm is applied to calculate the current states and transition states, as shown in Figure 2 . The algorithm can calculate all the possible transitions between states using a transition matrix. Finally, using these transition matrices, probabilities can then be calculated for each state.
MACA is very similar in structure to the Bayesian networks. Both models use Markov chains to determine the transitions. Bayesian Networks are more powerful than MBCs, since they can deal more than one event at a time.

```python
from transformers import pipeline
import os
import warnings
warnings.filterwarnings('ignore')
PWD=os.getcwd()
sa = pipeline("sentiment-analysis", top_k = None,
              model = "cardiffnlp/twitter-roberta-base-sentiment-latest",
              device = 0, padding = True, truncation = True,
              max_length = 512,
              model_kwargs={"cache_dir":PWD},
              verbose = False)
```

```python
print(sa("China slowdown weighs on revenue growth at internet giant Alibaba"))
```

```python
sentimental_class=sa("China slowdown weighs on revenue growth at internet giant Alibaba")
sentimental_class
```

```python
sentimental_class[0]
```

```python
sentimentality=[]
for x in sentimental_class[0]:
  sentimentality.append(x['score'])
  max_score=max(sentimentality)
  if x['score']==max_score:
    print(x['label'])
```
